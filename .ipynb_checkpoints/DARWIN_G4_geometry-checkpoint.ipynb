```python
###############
### Imports ###
###############

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib import gridspec            # to arrange the plots nicely
%matplotlib inline

'''
params = {
    'font.family': 'serif',
    etc.
}
plt.rcParams.update(params)
'''
plt.close()

import ROOT
```

```python
#def get_tree(filename):
#    f = ROOT.TFile(filename)
#    print(f)
#    f.cd('events')
#    print(f)
#    tree = f.Get('events').Get('events')
#    print(tree)
#    return tree

def get_position(filename):
    f = ROOT.TFile(filename)
    f.cd('events')
    tree = f.Get('events').Get('events')
    x=[]
    y=[]
    z=[]
    r=[]
    for entry in tree:
        x.append(entry.xp_pri)
        y.append(entry.yp_pri)
        z.append(entry.zp_pri)
        r.append(np.sqrt(entry.xp_pri*entry.xp_pri+entry.yp_pri*entry.yp_pri))
    return r, z, x, y

def plot(color, label, alpha=0.9):
    ax.scatter(r,z, color=color, edgecolor='none', marker='s', s=3, alpha=alpha, label=label)
    colors.append(color)
    return
```

```python
fig = plt.figure(figsize=(12,18))
ax = plt.subplot(111)
colors=[]

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/NeutronVeto_Scintillator.root')
plot('fuchsia', label='NeutronVeto_Scintillator')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/MuonVeto.root')
plot('darkmagenta', label='MuonVeto')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/Ti.root')
plot('black', label='Ti')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/Vacuum.root')
plot('gray', label='Vacuum')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/LXe.root')
plot('darkblue', label='LXe')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/GXe.root')
plot('mediumslateblue', label='GXe')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/PTFE_reflector.root')
plot('r', label='PTFE_reflector')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/PTFE_Pillar.root')
plot('lightcoral', label='PTFE_pillar')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/Cu_FSR.root')
plot('darkorange', label='Cu_FSR')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/SS304_topelectrodes_frame.root')
plot('g', label='SS304_topelectrodes_frame')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/SS304_cathode_frame.root')
plot('springgreen', label='SS304_cathode_frame')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/SS304_bottomscreen_frame.root')
plot('olive', label='SS304_bottomscreen_frame')

r, z, x, y = get_position('../DARWIN_Geometry/output_PR/PMT_sensorarray.root')
plot('gold', label='PMT_sensorarray')




leg = ax.legend(loc=1, bbox_to_anchor=[1.8,1.])
for color,text in zip(colors,leg.get_texts()):
    text.set_color(color)

plt.ylim(-4200,4200)
plt.xlim(0,4200)
plt.xlabel('R [mm]')
plt.ylabel('Z [mm]')
plt.grid()

plt.show()
```
